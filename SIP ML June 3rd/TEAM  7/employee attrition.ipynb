```python
x=data.iloc[:,4]
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn import cross_validation, neighbors
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train = y_train.astype('int')
y_train
```

```python
x_test
```

```python
y_test = y_test.astype('int')
y_test
```

```python
x_train = x_train.values.reshape((-1,1))
```

```python
y_train = y_train.values.reshape((-1,1))
```

```python
x_test = x_test.values.reshape((-1,1))
```

```python
y_test = y_test.values.reshape((-1,1))
```

```python
#KNN
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
print("Attrition : ")
print(y_test)
```

```python
print("\nAttrition : ")
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix

print("\nAccuracy score: %f" %(accuracy_score(y_test,y_pred) * 100))
print("ROC score : %f\n" %(roc_auc_score(y_test, y_pred) * 100))
print(confusion_matrix(y_test, y_pred)) 
```

```python
probas = clf.predict_proba(x_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.hist(probas, bins=20)
plt.title('Attrition Probabilities')
plt.xlabel('Probability')
plt.ylabel('# of Instances')
plt.xlim([0.5, 1.0])
plt.legend(y_test)
plt.show()
```

```python
#Evaluation Metrics
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
#Print cross validation score for kNN
clf_acc = cross_val_score(clf, x_train, y_train, cv=3, scoring="accuracy",  n_jobs=-1)
```

```python
clf_proba = cross_val_predict(clf, x_train, y_train, cv=3, method='predict_proba')
clf_scores = clf_proba[:, 1] # get the score of the positive class 
```

```python
clf_acc
```

```python
#Logistic Regression Model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr_acc = cross_val_score(lr, x_train, y_train, cv=3, scoring='accuracy', n_jobs=-1)
lr_proba = cross_val_predict(lr, x_train, y_train, cv=3, method='predict_proba')
lr_scores = lr_proba[:, 1]
```

```python
lr_acc
```

```python
#Predict on test data
y_pred=lr.predict(x_test)
```

```python
print("Attrition : ")
print(y_test)
```

```python
print("\nAttrition : ")
print(y_pred)
```

```python
#Accuracy score on Test and Train
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, confusion_matrix

print("\nAccuracy score: %f" %(accuracy_score(y_test,y_pred) * 100))
print("ROC score : %f\n" %(roc_auc_score(y_test, y_pred) * 100))
print(confusion_matrix(y_test, y_pred))
```
