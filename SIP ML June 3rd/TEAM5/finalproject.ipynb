```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_prices=pd.read_csv("CarPrice.csv")
```

```python
car_prices=car_prices.drop(['symboling','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation'],axis=1)   
```

```python
car_prices.head()
```

```python
car_prices=car_prices.drop(['curbweight','enginetype','cylindernumber','fuelsystem','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg'],axis=1)
```

```python
car_prices=car_prices.drop(['carheight'],axis=1)
```

```python
car_prices
```

```python
X=car_prices.iloc[:,0:-1].values
```

```python
X.shape
```

```python
y=car_prices.iloc[:,-1].values
```

```python
y=y.reshape(-1,1)
```

```python
y
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
from sklearn import linear_model as lm
model=lm.LinearRegression()
results=model.fit(X_train,y_train)
```

```python
accuracy=model.score(X_train,y_train)
print("Accuracy of the model:",accuracy)
```

```python
predictions=model.predict(X_test)
```
